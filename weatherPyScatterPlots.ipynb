```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import time
```

```python
matplotlib.rcParams['font.family'] = ['Poppins', 'Trebuchet MS']
```

```python
# file to load
file_to_load = 'weather_data/cities.csv'
weather_data_df = pd.read_csv(file_to_load)
weather_data_df.head()
```

```python
# get today's date in seconds
today = time.strftime('%x')
today
```

```python
lats = weather_data_df['Lat']
humidity = weather_data_df['Humidity']
max_temp = weather_data_df['Max Temp']
cloudiness = weather_data_df['Cloudiness']
wind_speed = weather_data_df['Wind Speed']


def plotDf(x, y, kolor):
    fig, ax = plt.subplots(figsize=(17, 12))
    plt.scatter(x, 
                y,
                edgecolor='#ccc', 
                linewidth=1.2, 
                marker='o', 
                alpha=0.8, 
                label='Citites', 
                s=200)
    plt.show();
```

```python
# y values
yValues = [humidity, max_temp, cloudiness, wind_speed]

# colors
humidityKolor, tempKolor, cloudinessKolor, windspeedKolor = ['tab:brown', 'tab:red', 'steelblue', 'tab:purple']

# titles
humidityTitle = 'Humidity'
tempTitle = 'Max Temperature'
cloudinessTitle = 'Cloudiness'
windspeedTitle = 'Wind Speed'

fig, ax = plt.subplots(figsize=(15, 7))
plt.scatter(lats,
            max_temp,
            color=tempKolor,
            edgecolors='black',
            linewidths=1,
            marker='o',
            alpha=0.7,
            label='Cities',
            s=150
           )
plt.title(f"City Latitude vs Max Temperature ({time.strftime('%x')})", fontsize=22, color=tempKolor)
plt.xlabel('Latitude', fontsize=17, color='steelblue')
plt.ylabel('Max Temperature (F)', fontsize=17, color='steelblue')
plt.xlim(-65, 85)
plt.ylim(-20, 115)
plt.tick_params(
    axis='both',
    direction='out',
    length=5,
    width=1,
    colors='slategray',
    pad=5,
    labelsize=14,
    grid_color='#bbb',
    labelrotation=0,
    grid_alpha=0.7,
    grid_linewidth=0.5
)
plt.grid(True)
plt.savefig('weather_data/lat-maxtemp.jpg')
plt.show();
```
